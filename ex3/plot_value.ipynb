```python
import sys, os
sys.path.insert(0, os.path.abspath(".."))
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt
import seaborn as sns

from common import helper as h
```

```python
# load the Q-value array
env_name = "CartPole-v0"

work_dir = Path().cwd()/'results'/f'{env_name}'
data = h.load_object(work_dir/'q_table_T1_1_b.pkl') # load q_table 
q_axis, q_table = data['axis'], data['q_table']
x_axis, th_axis = q_axis[0], q_axis[2]  #get the axis for x and \theta

discr = q_table.shape[0]
```

```python
########## Your code begins here. ##########
# Calculate the value function
values = np.max(q_table, axis=4)

# Plot the heatmap
values_array = values.mean(axis=(1,3))

ax = sns.heatmap(values_array, xticklabels=th_axis, yticklabels=x_axis)
plt.title('Heatmap')
ax.set_xlabel('theta')
ax.set_ylabel('x')
plt.show()

########## Your code ends here. ##########
```
